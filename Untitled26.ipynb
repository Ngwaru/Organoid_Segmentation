```python
import os
import cv2
import torch
import numpy as np
import pandas as pd
import torch.nn as nn
import torch.functional as F
from torch.utils.data import DataLoader, Dataset, random_split
```

```python
class MakeDataset(Dataset):
  def __init__(self, image_path, mask_path):
    super(MakeDataset).__init__()
    images = []
    masks = []
    for img_name in os.listdir(image_path):

      if img_name.endswith(".jpg"):
        try:
          img = cv2.resize(cv2.imread(f"{image_path}/{img_name}"), (128, 128))
          msk = cv2.resize(cv2.imread(f"{mask_path}/{img_name}"), (128, 128))
          img.shape
          msk.shape
          images.append(img)
          masks.append(msk)
        except Exception as e:
          print(e)
          print(img_name)

    self.data = torch.tensor(np.array(images), dtype=torch.float)/255
    self.masks = torch.tensor(np.array(masks), dtype=torch.float)/255

  def __getitem__(self, idx):

    return self.data[idx], self.masks[idx]

  def __len__(self):
    return len(self.data)
```
